```python
from langchain.chains.summarize import load_summarize_chain
from langchain.document_loaders import PyPDFLoader
from langchain import OpenAI, PromptTemplate
from langchain.document_loaders import Docx2txtLoader
from langchain.document_loaders import TextLoader
from langchain.docstore.document import Document
from langchain.text_splitter import CharacterTextSplitter
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.callbacks import get_openai_callback
from langchain.vectorstores import FAISS
from langchain.embeddings import OpenAIEmbeddings
from langchain.vectorstores import Milvus
```

```python
llm = OpenAI(temperature=0,openai_api_key=API_KEY)
```

SUMMARIZER

```python
def load_document(file_path):
    """ this function is used for loading the file """
    try:

        if file_path.endswith(".txt"):
            loader = TextLoader(file_path)
            pages=loader.load()
            logger.info("your text file is loaded")
        elif file_path.endswith(".pdf"):
            loader = PyPDFLoader(file_path)
            pages=loader.load_and_split()

            logger.info("your pdf file is loaded")
        else :
            loader = Docx2txtLoader(file_path)
            logger.info("your document file is loaded")
            pages=loader.load()
        return pages

    except :
        logger.error("unsupported files")
        
```

```python
pages=load_document("ChatGPTandAcademicResearch.pdf")
print(pages)
```

```python
text=""

for page in pages:
    text+=page.page_content

text=text.replace('\t',' ')
print(text)
```

```python
num_token=llm.get_num_tokens(text)
print(num_token)
```

```python
text_splitter = RecursiveCharacterTextSplitter(separators=["\n\n", "\n"],chunk_size=10000,chunk_overlap=500)
docs=text_splitter.create_documents([text])
```

```python
def split_docs():
    """ this function will split the documents into chunks"""
    #pages=loader.load_and_split()
    text=""
    for page in pages:
        text+=page.page_content
        
    text=text.replace('\t',' ')

    text_splitter = RecursiveCharacterTextSplitter(separators=["\n\n", "\n"],chunk_size=10000,chunk_overlap=10)

    splited_document=text_splitter.create_documents([text])
    num_splitted_docs=len(splited_document)
    num_tokens_first_doc = llm.get_num_tokens(splited_document[0].page_content)
    logger.info(num_splitted_docs)
    logger.info(num_tokens_first_doc)

    return splited_document
```

```python
splited_document=split_docs()
print(splited_document)
```

```python
prompt_template="""Write a concise summary in bullet point only of the following .
   {text}
BULLET POINT SUMMARY:"""

prompt=PromptTemplate(template=prompt_template,input_variables=["text"])
with get_openai_callback() as cb:
    summary=load_summarize_chain(llm,chain_type='map_reduce',map_prompt=prompt,combine_prompt=prompt)
    #output = summary.run([splited_document])
    output=summary({"input_documents":splited_document}, return_only_outputs=True)

    print(cb)

print(output)
```

```python
for i in output['output_text'].split('\n'):
    print(i)
```

```python
prompt_template="""Write a concise summary about {prospect}. If the information is not about {prospect}, exclude it from your summary.
Write a concise summary in bullet point only of the following .
   {text}
BULLET POINT SUMMARY:"""

prompt=PromptTemplate(template=prompt_template,input_variables=["text","prospect"])
with get_openai_callback() as cb:
    summary=load_summarize_chain(llm,chain_type='map_reduce',map_prompt=prompt,combine_prompt=prompt)
    output=summary({"input_documents":splited_document,"prospect":"Methodology"},return_only_outputs=True)

    print(cb)

print(output)
```
